```python
import torch

def tensor_basics():
    """
    Створіть функцію, яка:
    1. Створює два тензори розміром (3, 4) з випадковими числами
    2. Виконує елементне додавання, віднімання, множення
    3. Транспонує перший тензор
    4. Виконує матричне множення (після транспонування)
    5. Повертає всі результати
    """
    
    tensor1 = torch.rand((3, 4))
    tensor2 = torch.rand((3, 4))
    print(f"1 tensor: {tensor1}")
    print(f"\n2 tensor: {tensor2}")


    tensor_add = tensor1 + tensor2
    tensor_sub = tensor1 - tensor2
    tensor_mul = tensor1 * tensor2
    print(f"\naddition: {tensor_add}")
    print(f"\nsubtraction: {tensor_sub}")
    print(f"\nmultiplication: {tensor_mul}")


    tensor1_transposed = tensor1.T
    print(f"\ntransposed 1 tensor: {tensor1_transposed}")


    tensor_matmul = torch.matmul(tensor1_transposed, tensor2)
    print(f"\nmatrix multiplication: {tensor_matmul}")


results = tensor_basics()
```

```python
import torch  
 
def reshape_exercise():
    """
    Дано тензор розміром (2, 3, 4):
    1. Перетворіть його в (6, 4)
    2. Перетворіть його в (24,)
    3. Додайте новий вимір, щоб отримати (1, 2, 3, 4)
    4. Видаліть зайві виміри розміру 1

    Використайте reshape, flatten, unsqueeze, squeeze
    """
    tensor = torch.randn(2, 3, 4)
    print(f"Original tensor: shape={tensor.shape}\n{tensor}")

    reshape = tensor.reshape(6, 4)
    print(f"\nreshaped tensor (6,4): shape={reshape.shape}\n{reshape}")

    flatten = tensor.flatten()
    print(f"\nflattened tensor (24,): shape={flatten.shape}\n{flatten}")

    unsqueezed = tensor.unsqueeze(0)
    print(f"\nunsqueezed tensor (1,2,3,4): shape={unsqueezed.shape}\n{unsqueezed}")

    squeezed = unsqueezed.squeeze()
    print(f"\nSqueezed tensor: shape={squeezed.shape}\n{squeezed}")


result = reshape_exercise()
```

```python
import torch
 
def device_management():
    """
    1. Визначте доступний device (cuda або cpu)
    2. Створіть тензор на CPU
    3. Якщо є GPU - перемістіть на GPU
    4. Виконайте просту операцію (наприклад, квадрат)
    5. Поверніть результат на CPU

    Код має працювати незалежно від наявності GPU!
    """

    if torch.cuda.is_available():
        device = "cuda"
    else:
        device = "cpu"

    print(f"Using device: {device}")
    
    tensor_cpu = torch.randn(3, 3)
    print(f"\nOriginal tensor on CPU: {tensor_cpu}")

    tensor_gpu = tensor_cpu.to(device)
    print(f"Tensor moved to {device}: {tensor_gpu}")

    tensor_squared = tensor_gpu ** 2
    print(f"Squared tensor on {device}: {tensor_squared}")

    tensor_cpu = tensor_squared.to("cpu")
    print(f"Result moved back to CPU: {tensor_cpu}")


result = device_management()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_linear_data(n_samples=100, n_features=5, noise=0.1):
    """
    Згенеруйте дані для лінійної залежності:
    y = X @ w + b + шум

    1. Створіть випадкові ваги w та зміщення b
    2. Згенеруйте випадкові входи X
    3. Обчисліть y за формулою
    4. Додайте гаусівський шум

    Поверніть: X, y, true_w, true_b
    """

    true_w = np.random.randn(n_features)  
    true_b = np.random.randn()            

    # input x
    X = np.random.randn(n_samples, n_features) 

    y = X @ true_w + true_b

    y += np.random.normal(0, noise, size=y.shape)

    return X, y, true_w, true_b
    

X, y, w, b = generate_linear_data(200, 2, 0.1)


print("True weights:", w)
print("True bias:", b)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1) ## 1 rows 2 columns, first subplot
plt.scatter(X[:,0], y, alpha=1, color='blue') ## alpha - transparency
plt.xlabel("feature 1") ## X name 
plt.ylabel("y")         ## Y name

plt.subplot(1,2,2) 
plt.scatter(X[:,1], y, alpha=1, color='yellow')
plt.xlabel("feature 2")
plt.ylabel("y")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import torch
import time

def compare_operations():
    """
    Порівняйте час виконання для матричного множення 1000x1000:
    1. Два вкладені цикли Python (для маленької матриці 10x10!)
    2. NumPy операція
    3. PyTorch CPU
    4. PyTorch GPU (якщо доступно)

    Виведіть час кожного методу
    """
    size = 1000


    ## NumPy
    a_np = np.random.randn(size, size)
    b_np = np.random.randn(size, size)

    start = time.time()
    c_np = a_np @ b_np
    numpy_time = time.time() - start


    ## Python loops 
    size_10 = 10

    a_py = np.random.randn(size_10, size_10)
    b_py = np.random.randn(size_10, size_10)
    
    start = time.time()
    c_py = np.zeros((size_10, size_10)) # matrix

    for i in range(size_10):
        for j in range(size_10):
            for k in range(size_10):
                c_py[i, j] += a_py[i, k] * b_py[k, j]
    loops_time = time.time() - start


    ## PyTorch CPU
    cpu_device = torch.device("cpu")
    a_torch = torch.randn(size, size, device=cpu_device)
    b_torch = torch.randn(size, size, device=cpu_device)

    start = time.time()
    c_torch = a_torch @ b_torch
    torch_cpu_time = time.time() - start


    ## PyTorch GPU
    gpu_device = torch.device("cuda")
    a_torch = torch.randn(size, size, device=gpu_device)
    b_torch = torch.randn(size, size, device=gpu_device)
    
    torch.cuda.synchronize()  # for correct timing GPU
    start = time.time()
    c_torch = a_torch @ b_torch
    torch.cuda.synchronize()
    torch_gpu_time = time.time() - start

    print("Time with Python loops (10x10):", loops_time)
    print("Time with NumPy (1000x1000):", numpy_time)
    print("Time with PyTorch CPU (1000x1000):", torch_cpu_time)
    print("Time with PyTorch GPU (1000x1000):", torch_gpu_time)

    
compare_operations()
```
